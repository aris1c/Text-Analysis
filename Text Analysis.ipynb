##Regular Expression (RegExp)

```python
import re
```

```python
#search Data using RegExp (search)
re.search(r'\b\w{5}\b','Andri and Fatia')
```

```python
#search Data using RegExp (match)
re.match(r'\b\w{5}\b','Andri and Fatia')
```

```python
#search Data using RegExp (findall)
re.findall(r'\b\w{5}\b','Andri and Fatia')
```

```python
#replace data
s = "User's mobile number is 1234567890"
s_hide = re.sub('\d','*',s)
print(s_hide)
```

```python
#Data validator
user_name = input("Please enter your name: ")

while not re.match("^[A-Za-z]+$",user_name):
  print("Error! Make sure you only use letter in your name")
  user_name = input("Please enter your name: ")

print("Hello!"+user_name)
```

```python
from fuzzywuzzy import fuzz
```

```python
text1 = "The quick brown fox jumps over the lazy dog"
text2 = "The quick brown fox jumps over the lazy dog."

similarity = fuzz.ratio(text1,text2)
print(similarity)
```

```python
similarity = fuzz.partial_ratio(text1,text2)
print(similarity)
```

```python
str1 = "My name is Ali"
str2 = "Ali is my name"
Ratio = fuzz.ratio(str1.lower(),str2.lower())
print(Ratio)
```

```python
str1 = "My name is Ali"
str2 = "My name is Ali Abdaal"
print(fuzz.partial_ratio(str1.lower(),str2.lower()))
```

```python
str1 = "My name is Ali"
str2 = "Ali is my name"
print(fuzz.token_sort_ratio(str1.lower(),str2.lower()))
```

```python
str1 = "My name is Ali"
str2 = "Ali is my name name"
print(fuzz.token_sort_ratio(str1.lower(),str2.lower()))
print(fuzz.token_set_ratio(str1.lower(),str2.lower()))
```

```python
strA = "Read the sentence - My name is Ali"
strB = "My name is Ali"
print(fuzz.token_set_ratio(strA,strB))
```

```python
from fuzzywuzzy import process
query = "My name is Ali"
choices = ["My name Ali", "My name is Ali", "My Ali"]
#get a list of matcher ordered by score, default limit to 5
process.extract(query,choices)
```

```python
import pandas as pd

# creating a dataframe
dict_one = {
    "country" : ["England","Wales","Scotland","united kingdom","Northern Ireland"],
    "population_in_millions": [55.98,5.45,3.14,67.33,1.89]
}

dict_two = {
    "country" : ["Northern Iland","Wles","Scotland","England","United K."],
    "GDP_per_capita": [24900,23882,37460,45101,46510.28]
}

existing_data = pd.DataFrame(dict_one)
exported_data = pd.DataFrame(dict_two)

print(existing_data,exported_data, sep="\n\n")
```

```python
data = pd.merge(existing_data,exported_data,on="country",how="left")
print(data)
```

##Sentiment Analysis

```python
from textblob import TextBlob

text1 = "The movie was so awesome."
text2 = "The food here testes terrible."

#Determining the Polarity
p_1 = TextBlob(text1).sentiment.polarity
p_2 = TextBlob(text2).sentiment.polarity

#Determining the subjectivity
s_1 = TextBlob(text1).sentiment.subjectivity
s_2 = TextBlob(text2).sentiment.subjectivity

print("Polarity of Text 1 is", p_1)
print("Polarity of Text 2 is", p_2)
print("subjectivity of text 1 is", s_1)
print("subjectivity of text 2 is", s_2)
```

1.0 +1 itu sentiment positif
-1.0 -1 itu sentiment negatif

##Name Entity Recognition (NER)

```python
import spacy
from spacy import displacy

NER = spacy.load("en_core_web_sm")
```

```python
raw_text = "The Indian Space Research Organization or is the national space agency of India, headquartered in Bengaluru.  It operates under Departement of Space with is directly over price BTC WILL go to 1M Rp and will be the crypto currency the best of the best"
```

```python
raw_text2 = "The Mars Orbited Mission(MOM), informally known as Mangalyaan, was launched into Earth orbit on 5 November 2013 by the Indian Space Research Organization (ISRO) is Elon Musk create rocket to the moon, and he put Doge Coin into it as Doge 1 Mission to the moon, master roshi is the one who created btc as satoshi masamoto great! currency"
```

```python
text1 = NER(raw_text)
```

```python
text2 = NER(raw_text2)
```

```python
for word in text1.ents:
  print(word.text,word.label_)
```

```python
for word in text2.ents:
  print(word.text,word.label_)
```

```python
displacy.render(text2,style="ent",jupyter=True)
```
